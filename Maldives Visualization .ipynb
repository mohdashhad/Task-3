```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load Excel data
df = pd.read_excel(r"C:\Users\Hp\Downloads\Maldives_Final_Keyword_Report.xlsx", sheet_name='Sheet1')

# Step 2: Count occurrences of each Source URL
source_counts = df['Source URL'].value_counts()

# Step 3: Plot the pie chart
plt.figure(figsize=(10, 10))
wedges, texts, autotexts = plt.pie(
    source_counts.values,
    autopct='%1.1f%%',
    startangle=90,
    textprops=dict(color="black", fontsize=9)
)

# Step 4: Add legend (cleaner than labels on pie)
plt.legend(
    wedges,
    source_counts.index,
    title="Source URL",
    bbox_to_anchor=(1.05, 0.8),
    loc="upper left",
    fontsize='small'
)

# Step 5: Title and format
plt.title('Count of Source URL')
plt.axis('equal')  # Equal aspect ratio for a perfect circle

# Optional: Save the chart as PNG
# plt.savefig("source_url_pie_chart.png", bbox_inches='tight', dpi=300)

# Step 6: Show the plot
plt.show()
```
